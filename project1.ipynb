```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv


path="/content/drive/MyDrive/project.csv"
electricity=pd.read_csv(path)
print(electricity)
print(electricity.shape)
print(electricity.head(5))
print(electricity.tail(4))
print(electricity.info())
print(electricity.describe())
print("Column Data")
print(electricity.columns)
print(electricity[electricity['State'].str.startswith('T')])
print(electricity[electricity['State'].str.endswith('A')])
rcb=electricity[electricity['State']=='Tripura']

print(len(rcb))
print(electricity.sort_values(by='State'))
print(np.max(electricity['Consumption']))
print(np.min(electricity['Consumption']))
print(np.mean(electricity['Consumption']))
print(np.median(electricity['Consumption']))
print(np.std(electricity['Consumption']))

state1_data=electricity[electricity['State']=='Andhra']
print("Andhra Data")
print(state1_data)
state2_data=electricity[electricity['State']=='Tamil Nadu']
print("Tamil Nadu Data")
print(state2_data)
state3_data=electricity[electricity['State']=='Kerala']
print("Kerala Data")
print(state3_data)
state4_data=electricity[electricity['State']=='Karnataka']
print("Karnataka Data")
print(state4_data)
state5_data=electricity[electricity['State']=='Maharashtra']
print("Maharashtra Data")
print(state5_data)

fig, axes = plt.subplots(2)
axes[0].plot(state1_data['Date'], state1_data['Consumption'])
axes[1].plot(state2_data['Date'], state2_data['Consumption'],color="red")
plt.show()


state1='Andhra'
state2='Tamil Nadu'

fig1, axes1 = plt.subplots(2)
axes1[0].bar(state1_data['Date'], state1_data['Consumption'])
axes1[1].bar(state2_data['Date'], state2_data['Consumption'])
plt.show()

labels=[state1,state2]
state1_total=electricity[electricity['State']==state1]['Consumption'].sum()
state2_total=electricity[electricity['State']==state2]['Consumption'].sum()
fig2, axes2 = plt.subplots(1)
energy=[state1_total,state2_total]
axes2.pie(energy,labels=labels,autopct='%1.1f%%')


```
