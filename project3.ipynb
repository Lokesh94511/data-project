```python
import pandas as pd
from pandas import read_csv

path="/content/drive/MyDrive/rawdata.csv"
df=pd.read_csv(path)
new_df=df.dropna()
print(new_df.to_string())
print(new_df)
print(df)
#replace empty values with 130
new_df=df.fillna(130)
print(new_df.to_string())
#replace only for specified colums
new_df=df["Calories"].fillna(130)
print(new_df.to_string())
new_df=df["Pulse"].fillna(80)
print(new_df.to_string())
#replace using mean,median,mode
x = df["Duration"].mean()
df["Duration"].fillna(x)
print(new_df.to_string())
#median
y=df["Pulse"].median()
df["Pulse"].fillna(y)
print(new_df.to_string())
#mode
z=df["Maxpulse"].mode()
df["Maxpulse"].fillna(z)
print(new_df.to_string())
#fixing wrong data
df.loc[7,'Duration']=45
print(df.to_string())
#looping
for x in df.index:
  if df.loc[x,"Duration"]>120:
    df.loc[x,"Duration"]=120
print(df.to_string())
#revoming duplicates
path="/content/drive/MyDrive/rawdata.csv"
df=pd.read_csv(path)
print(df.duplicated())
df.drop_duplicates(inplace=True)
df.corr()


```
